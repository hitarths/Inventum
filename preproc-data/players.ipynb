```python
import pandas as pd
import numpy as np
from numpy.linalg import norm
import pickle
from sklearn import preprocessing
```

```python
datadir = '../real-life-datasets/'
fname = 'players'
fdata = fname + '.csv'
```

```python
df = pd.read_csv(datadir+fdata, encoding="latin-1", delimiter=';')
names = df['Player'] + df['Year'].astype(str)
df = df.drop(['Player','Year','Rk','Pos','Tm','GS','FGA','3PA','2PA','FTA','TRB'], 
        axis=1)
df = df.apply(pd.to_numeric, errors='coerce')
df
```

```python
items = df.dropna()
items.reset_index(drop=True, inplace=True)
```

```python
items
```

```python
items.dtypes
```

```python
# Normalize

items_normalized = items.copy()  # Create a copy of the DataFrame

# Normalize numerical columns (assuming all columns are numeric)
for column in items_normalized.columns:
    if items_normalized[column].dtype in [int, float]:  # Check if the column contains numeric data
        min_val = items_normalized[column].min()
        max_val = items_normalized[column].max()
        items_normalized[column] = (items_normalized[column] - min_val) / (max_val - min_val)

items_normalized
```

```python
import os

with open(os.path.expanduser(datadir+fname+'.pkl'), 'wb') as fout:
    pickle.dump([items_normalized,names], fout)
```
