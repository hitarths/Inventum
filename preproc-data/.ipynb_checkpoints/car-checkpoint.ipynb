```python
import pandas as pd
import numpy as np
from numpy.linalg import norm
import pickle
from sklearn import preprocessing
```

```python
pd.set_option('display.max_columns', 500)  # or 1000
```

```python
datadir = '~/Downloads/datasets-interactive-top-tuples/'
fname = 'car'
fdata = fname + '.csv'
```

```python
df = pd.read_csv(datadir+fdata, delimiter=',') # na_values='NaN'
```

```python
df
```

```python
df.isna().sum()
```

```python
df = df.drop(['name','vehicleConfiguration','engineName','engineDisplacement','date','location','link','parse_date',
               # 'year' this line due to NaN
             ],
        axis=1)
```

```python
df = df.dropna()
df.shape
```

```python
df['brand'].unique().shape, df['bodyType'].unique().shape, df['color'].unique().shape, df['fuelType'].unique().shape, df['transmission'].unique().shape
```

```python
df = df.drop(['brand','color'], axis=1)
```

```python
for c in ['bodyType','fuelType','transmission']:
    df1 = pd.get_dummies(df[c])
    df = pd.concat([df, df1], axis=1)

df = df.drop(['bodyType','fuelType','transmission'], axis=1)
df.shape
```

```python
df['year'] = 2022 - df['year']
```

```python
df.dtypes.tolist()
```

```python
df
```

```python
names = None
items = df

x = items.values
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
x_scaled = x_scaled / norm(x_scaled, axis=1)[:, None]
items = pd.DataFrame(x_scaled, columns=items.columns)
```

```python
items
```

```python
import os

with open(os.path.expanduser(datadir+fname+'.pkl'), 'wb') as fout:
    pickle.dump([items,names], fout)
```
